* Zoek de oudste en jongste persoon uit je dictionary. Geef ook de leeftijd weer.

````python
import datetime

naam_en_geboortejaar = input("Geef naam en geboortejaar in: ")
gegevens = {}
while naam_en_geboortejaar != "stop":
    naam = naam_en_geboortejaar.split(",")[0]
    leeftijd = int(naam_en_geboortejaar.split(",")[1])

    gegevens[naam] = leeftijd
    naam_en_geboortejaar = input("Geef naam en leeftijd in: ")
print(gegevens)

oudste = None
jongste = None
for mensen in gegevens:
    oudste = mensen if  not oudste or gegevens[mensen] < gegevens[oudste] else oudste
    jongste = mensen if  not jongste or gegevens[mensen] > gegevens[jongste] else jongste

    jaar = datetime.date.today().year
    oudste_leeftijd =  jaar - gegevens[oudste]
    jongste_leeftijd =  jaar - gegevens[jongste]
print(oudste)
print(jongste)
print(oudste_leeftijd)
print(jongste_leeftijd)

````

* Schrijf een functie _verwijder\_lege_lijsten()_ die dit soort dictionary als input neemt en de paren met een lege lijst als value hieruit verwijdert. Vb.:

```
num['n4'] = []
verwijder_lege_lijsten(num)
>> {'n1': [2, 3, 1], 'n2': [5, 1, 2, 7], 'n3': [3, 4]}

```

````python
num = {'n1': [2, 3, 1], 'n2': [5, 1, 2, 7], 'n3': [3, 4]}
def sorteer_lijsten(dictionary):
    for keys in dictionary:
        dictionary[keys].sort(reverse=True)
    return dictionary
print(sorteer_lijsten(num))

num = {'n1': [2, 3, 1], 'n2': [5, 1, 2, 7], 'n3': [3, 4]}
def lengte_lijsten(dictionary):
    for keys in dictionary:
        dictionary[keys] = len(dictionary[keys])
    return dictionary
print(lengte_lijsten(num))

num = {'n1': [2, 3, 1], 'n2': [5, 1, 2, 7], 'n3': [3, 4]}
def gemid_lijsten(dictionary):
    for keys in dictionary:
        dictionary[keys] = sum(dictionary[keys])/len(dictionary[keys])
    return dictionary
print(gemid_lijsten(num))

num = {'n1': [2, 3, 1], 'n2': [5, 1, 2, 7], 'n3': [3, 4]}
num['n4'] = []
def verwijder_lege_lijsten(dictionary):
    verwijderen = list()
    for keys in dictionary:
        if not dictionary[keys]:
            verwijderen.append(keys)
    
    for keys in verwijderen:
        dictionary.pop(keys)
    return dictionary
print(verwijder_lege_lijsten(num))


````

Gebruik het Romeinse cijfer als key, en ons natuurlijk cijfer als value.

Schrijf nu een functie _roman\_calculator()_ die deze dictionary als input neemt, samen met een Romeins getal. Je functie zet deze om naar een natuurlijk getal. Hou hierbij rekening met volgende rekenregel:
- De waarden van de losse symbolen worden bij elkaar opgeteld, behalve als een symbool met een lagere waarde vóór een symbool met een hogere waarde staat. In dat laatste geval wordt de lagere waarde ervan afgetrokken. Voor de rest is de volgorde van de getallen van hoog naar laag.

Voorbeelden:
```
"VIII" geeft 8
"IIX" geeft 8
"IL" geeft 49
"XXXXVIIII" geeft 49
"XLV" geeft 45
"VL" geeft 45
```


````python
romeinse_cijfers = {"I" : 1, "V" : 5, "X" : 10, "L" : 50, "C" : 100, "D" : 500, "M" : 1000}

def roman_calculator(romeins_getal):
    romeins_getal = romeins_getal[::-1]
    getal = 0
    opgeslagen_getal = 0
    for i in range(len(romeins_getal)):

        if romeinse_cijfers[romeins_getal[i]] >= opgeslagen_getal:
            getal += romeinse_cijfers[romeins_getal[i]]
            opgeslagen_getal = romeinse_cijfers[romeins_getal[i]]
        else:
            getal -= romeinse_cijfers[romeins_getal[i]]
    return getal
print(roman_calculator("VL"))
````
