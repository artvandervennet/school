Schrijf een Python functie _count\_letters()_ die een string accepteert en het aantal hoofdletters en kleine letters in de string telt. Je functies geeft dus 2 resultaten terug.

````python
def count_tellers(string = ""):
    klein = 0
    groot = 0
    for letters  in string:
        if letters.islower():
            klein += 1
        elif letters.isupper():
            groot += 1
    return klein, groot

klein, groot = count_tellers("kawgyEFASDWdkua")
print(klein, groot)
````

### Oefening 8

Schrijf een Python functie _seperate()_ die een getal $n$ als input neemt en die alle even en alle oneven getallen kleiner dan $n$ scheidt in twee verschillende lijsten.

````python
def seperate(getal):
    oneven = []
    even = []
    for i in range(getal):
        if i % 2: #oneven
            oneven.append(i)
        else:
            even.append(i)
    return oneven, even

oneven, even = seperate(50)
print(oneven, even)
````

### Oefening 9

Schrijf een Python-functie die alle priemgetallen kleiner dan een gegeven getal $n$ teruggeeft in een lijst

````python
def is_prime(n):
    for i in range(2,n):
        if (n%i) == 0:
            return False
    return True

def prime_tellen(getal):
    prime = []
    for i in range(getal):
        if is_prime(i):
            prime.append(i)
    return prime

prime = prime_tellen(50)
print(prime)
````

### Oefening 13

Schrijf een programma dat studentengegevens kan beheren. Je programma zal bij aanvang vragen om een aantal studentengegevens in te geven, namelijk de naam van de student, zijn klasgroep (1ICT1,1EO1,1ICT2, 1EO2, enz ...) en zijn behaalde score op een test . Deze gegevens zal je programma intern opslaan in een rij van tuples. Wanneer al deze gegevens ingelezen en opgeslaan zijn, zal je programma een overzicht uitprinten van de gemiddelde score per klasgroep.

Bijvoorbeeld :

1ICT1 : 14 / 20

1ICT2 : 13 / 20 .....

Let op : laat je programma herhaaldelijk vragen of er nog een student toegevoegd moet worden. Indien niet (gebruiker typt nee), dan pas worden de gemiddelden berekend en uitgeprint.

Verdeel het probleem in subtaken en maak gebruik van functies.

````python
def vraag_gegevens():
    '''vraagt om alle gegevens in te geven van elke leerling, returned een array met per leerling een array [naam, klas, score]'''
    gegevens = []
    i = 0
    vraag = True if input("wil je nog een gebruiker toegoegen: ") != "nee" else False
    while vraag:
        gegevens.append([])
        gegevens[i].append(input("geef een naam: "))
        gegevens[i].append(input("geef een klas: "))
        gegevens[i].append(input("geef een score: "))
        i += 1
        vraag = True if input("wil je nog een gebruiker toegoegen: ") != "nee" else False
    return gegevens

def punten_per_klas(gegevens):
    '''returned een array met alle klassen en een array met respectievelijk alle punten per klas'''
    klasgroepen = []
    punten_in_klasgroepen = []
    for leerlingen in gegevens:
        klasgroep = leerlingen[1]
        punten = leerlingen[2]
        if klasgroep not in klasgroepen:
            klasgroepen.append(leerlingen[1])
            punten_in_klasgroepen.append([])
        punten_in_klasgroepen[klasgroepen.index(klasgroep)].append(int(punten))
    return klasgroepen, punten_in_klasgroepen

def gemiddeldes_per_klas(klasgroepen, punten_in_klasgroepen):
    '''berekend de de gemiddeldes per klas en zet deze respectievelijk met de klasgroepen in een array'''
    gemiddeldes_klassen = []
    for klassen in klasgroepen:
        gemiddeldes_klassen.append(sum(punten_in_klasgroepen[klasgroepen.index(klassen)]) / len(klasgroepen))
    return gemiddeldes_klassen


gegevens = vraag_gegevens()
klasgroepen, punten_in_klasgroepen = punten_per_klas(gegevens)
gemiddeldes = gemiddeldes_per_klas(klasgroepen, punten_in_klasgroepen)

for klasgroep in range(len(klasgroepen)):
    print(f"{klasgroepen[klasgroep]}: {gemiddeldes[klasgroep]}/20")
````
