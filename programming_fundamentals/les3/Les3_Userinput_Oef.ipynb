### Oefening 3

4. Converteer de geboortedatum naar het formaat JJJJ-MM-DD.
5. Print een zin die zegt: "Hallo [geformatteerde naam], u bent geboren op [geformatteerde geboortedatum]."

Voorbeeld:
```
1. Gebruiker voert in: "eVert-jAn"
```

Vraag de gebruiker om een geldig bankrekeningnummer in te geven. Opgelet, een geldig bankrekeningnummer wordt voorafgegaan door een IBAN code, bvb IBAN BE54 1234 5678 9012.
Print daarna de IBAN code en het eigenlijke rekeningnummer afzonderlijk af.

Deel vervolgens het getal, gevormd door de eerste tien cijfers door 97. De rest van deze deling moet gelijk zijn aan het controle getal, gevormd door de laatste 2 cijfers. Ga dit na.

```
Geef je rekeningnummer met IBAN code:IBAN BE54 1234 5678 9002
De IBAN code is: BE54
Het rekeningnummer is: 1234 5678 9002
1234567890
Het controle getal is: 2
```

`````python
rekeningnummer_IBAN = input("geef een geldig bankrekeningnummer in: ")

print(f"de IBAN code is: {rekeningnummer_IBAN.split(' ')[1]}")

print(f"het rekeningnummer is: {rekeningnummer_IBAN[10:]}")

controlegetal = int(rekeningnummer_IBAN[10:-2].replace(' ','')) % 97

print(f"het controlegetal is: {controlegetal}")

print(int(rekeningnummer_IBAN[-2:]) == controlegetal)

`````

 ### Oefening 10
 Schrijf een programma dat op basis van een ingelezen achternaam en voornaam van een klant een klantcode genereert op volgende manier :

- De key begint steeds met de prefix ”ODISEE”
- Het volgende karakter bepaal je als volgt : neem de decimale waarde van de derde letter van de voornaam trek hier exact 32 vanaf en zet terug om naar een letter.
- Voor het volgende karakter neem je de laatste letter van de familienaam en zet deze om naar een hoofdletter.
- Tenslotte eindig je de code door het volgende: neem de eerste letters van de voor- en achternaam. Vermenigvuldig hun numerieke waarden en dit getal plak je achteraan de code.

Merk op, de voornaam moet minstens bestaan uit 3 karakters, de achternaam kan samengesteld zijn bvb. _Van de Winkel_ beschouw dit als 1 woord, met spaties.

Dit geeft het volgende voor de naam : _Jan Janssens_:
- De derde letter van 'Jan' = 'n'. Hiervan nemen we n - 32 ofwel 110 - 32 = 78 wat volgens de ASCII tabel de letter 'N' geeft.
- Het volgende karakter is eveneens 'S'
- Het produkt van 'J' met zichzelf hier geeft : 74 x 74 = 5476

De code voor Jan Janssens is dan als volgt : ODISEENS5476

`````python
voornaam = input("geef je voornaam: ")
achternaam = input("geef je achternaam: ")

print(ord(voornaam[2]))

klantcode = f"ODISEE{ chr(ord(voornaam[2]) - 32) }{achternaam[-1].upper()}{ord(voornaam[0])*ord(achternaam[0])}"

print(klantcode)
`````
