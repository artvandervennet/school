### Oefening 9

Schrijf een programma dat het volume van een cilinder berekent. De formule voor het volume is
$$
V = \Pi \times R^2 \times h
$$

waarbij R de straal van de cilinder is (in te geven via het klavier) en h de hoogte (ook in te geven via het klavier).

Voorzie een gepaste uitvoer waarbij je het resultaat afrondt op 2 cijfers na de komma.

```python
r = int(input("geef een straal in: "))
h = int(input("geef een hoogte in: "))

V = math.pi * (r**2) * h

print(V)
print(round(V,2))
```

### Oefening 14

Schrijf een Python-script dat de gebruiker vraagt om het aantal gereden kilometers en het aantal liter gebruikte brandstof in te voeren. Het script moet vervolgens het brandstofverbruik per 100 km en de kostprijs per 100 km berekenen. De huidige benzineprijs bedraagt bedraagt 2,1050 euro/l. Neem deze waarde op als constante in je programma. Print alle berekeningen netjes uit tot 2 cijfers na de komma.

```python
gereden_km = float(input("hoeveel km heb je gereden: "))
liters_verbruikt = float(input("hoeveel liters heb je verbruikt: "))

KOSTPRIJS_PER_L= 2.1050

brandstofverbruik = (liters_verbruikt / gereden_km) * 100
kostprijs = brandstofverbruik * KOSTPRIJS_PER_L

print(f"je verbruikte gemmideld {brandstofverbruik : .2f}l per 100km")
print(f"je betaalde gemmideld {kostprijs : .2f} euro per 100km")

```

### Oefening 15
Schrijf een programma dat je toelaat om in te schatten of je je E-reader
moet herladen nog voor het beëindigen van je boek. Vraag de gebruiker
vooraf hoeveel pagina's hij momenteel al gelezen heeft. Vraag hierbij ook
wat de leestijd in minuten is die hij hieraan spendeerde. Ga er vanuit dat je Ereader een batterijtijd heeft van 9,5 uur. Neem deze waarde op als constante
in je programma.
Vraag vervolgens hoeveel resterende pagina's er nog moeten gelezen worden
en bereken dan de resterende leestijd die hiervoor nodig zal zijn. Geef je
antwoord terug in minuten en rond je resultaat af op 2 cijfers na de komma.
Pols vervolgens wat het huidige percentage is van de batterij van je E-reader.
Bereken hiermee de batterijtijd die je nog rest. Geef ook dit resultaat terug in
minuten en rond opnieuw af op 2 cijfers na de komma.

```python
aantal_paginas_gelezen = int(input("hoeveel pagina's heb je  al gelezen: "))
aantal_minuten_gelezen = float(input("hoelang heb je daarover gedaan (in min): "))
aantal_paginas_over = int(input("hoeveel pagina's zijn er nog over: "))
BATTERIJTIJD_PER_UUR = 9.5

batterijtijd_per_minuten = BATTERIJTIJD_PER_UUR*60
leestijd_nodig = (aantal_minuten_gelezen/aantal_paginas_gelezen) * aantal_paginas_over
print(f"je zal nog {leestijd_nodig} minuten lezen")

percentage_ereader = int(input("hoeveel percent batterij heeft de e-reader nog over: "))
minuten_over = batterijtijd_per_minuten * ((100 - percentage_ereader) / 100)
print(f"je e-reader heeft nog {minuten_over : .2f}min over")
```

### Oefening 16

Schrijf een programma dat de gemiddelde afdruksnelheid en het inktverbruik
van een standaard inktJet printer kan berekenen. De printer in kwestie print
gemiddeld 15 pagina's uit in 33 seconden. Voorzie een gepaste constante om
de printersnelheid op te nemen in je programma.
Vraag vervolgens hoeveel pagina's de gebruiker wil afprinten. Bereken met
deze input hoelang de printer erover doet om het aantal ingegeven pagina's
van de gebruiker uit te printen (in seconden).

Ga er vanuit dat je met een vol inktpatroon, dat ongeveer 19 ml inkt bevat,
gemiddeld 500 A4'tjes kan uitprinten. Neem ook al deze
cijfers op als constanten in je programma. Bereken nu ook voor het ingegeven aantal pagina's het inktverbruik (in ml).
Rond steeds je resultaten af tot op 2 cijfers na de komma.

```python
PRINTSNELHEID_PER_SEC = 15/33
hoeveel_paginas_printen = int(input("hoeveel pagina's wil je printen"))

VOL_INKTPATROON = 19
AANTAL_PRINTER_VOL = 500

inktverbruik = (VOL_INKTPATROON/AANTAL_PRINTER_VOL) * hoeveel_paginas_printen

printer_duur = hoeveel_paginas_printen / PRINTSNELHEID_PER_SEC

print(printer_duur)
print(f"de printer zal daarvoor {inktverbruik}ml inkt verbruiken")

```
