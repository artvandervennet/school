### Oefening 1

Schrijf een klasse _MeetGegevens_. Deze klasse bevat volgende methodes en data:

- Een constructormethode waarbij je de meetgegevens kan meegeven als parameter via een lijst van positieve integer getallen.
- Een methode _zijn\_alle\_elementen\_verschillend()_ die een boolean teruggeeft.
- Een methode _bepaal\_kleinste\_positief\_verschil()_ die een int waarde teruggeeft. Ga hierin na welke het kleinste verschil is tussen de meetgegevens en geef deze waarde terug.
- Een methode \_\_\_str\_\_()_ die een string opbouwt met hierin alle meetgegevens.

Schrijf nu een testprogramma. Maak eerst een lijst van postieve getallen aan. Maak hiermee een object van de klasse MeetGegevens. Via de bovenstaande methoden,  kan het programma nu controleren of deze waarden allemaal onderling verschillend zijn. Als dat het geval is, moet bepaald worden welke twee meetwaarden het dichtst bij elkaar liggen en het positieve verschil hiertussen moet uitgeschreven worden.

````python
class MeetGegevens:
    def __init__(self, gegevens):
        self.gegevens = gegevens
    
    def zijn_alle_elementen_verschillend(self):
        return len(self.gegevens) == len(set(self.gegevens))

    def bepaal_kleinste_positief_verschil(self):
        gegevens = sorted(self.gegevens)
        verschillen = []
        for i in range(len(self.gegevens) - 1):
            verschillen.append(gegevens[i+1] - gegevens[i])
        return min(verschillen)
    
    def __str__(self):
        return str(self.gegevens)

gegevens = [1,3,5,7,6,3,1]
meetgegevens = MeetGegevens(gegevens)
print(meetgegevens)
print(meetgegevens.zijn_alle_elementen_verschillend())
print(meetgegevens.bepaal_kleinste_positief_verschil())
````

### Oefening 2

Schrijf een klasse _LoopWedstrijd_. Deze zal de prestaties van verschillende atleten voor een bepaalde wedstrijd bijhouden en analyseren. Deze klasse bevat de volgende methodes en data.

- Een constructormethode die de naam van de wedstijd als input krijgt, en optioneel ook al enkele namen van de atleten in een lijst alsook een lijst met hun looptijden.

- Een methode _geef_winnaar()_ die nagaat welke atleet de beste tijd liep en de naam van deze atleet teruggeeft als resultaat. Bij ex aequo geef je een lijstje van winnaars terug.

- Een methode _voeg\_prestatie\_toe()_ die een atleet en zijn prestatie als parameter neemt en deze nieuwe informatie toevoegt aan de gegevens van je object. Let op, je kan geen dubbels toelaten, wanneer een atleet al in de gegevens van het object zit dan overschrijf je zijn oude tijd met zijn nieuwe.

- Een methode \_\_\_str\_\_()_ die een string opbouwt met hierin alle gegevens van de atleten en hun tijden.

Maak nu een object van type _LoopWedstrijd_ met de naam "OdiseeRun" voeg hierbij volgende atleten en tijden toe:
```
deelnemers: Peter, Kristien, Thomas, Evert-Jan
tijden: 12.5 14.8 13.2 14.5
```
Geef de voorlopige winnaar van deze wedstrijd:
```
en het goud gaat naar ...: Peter
```
Voeg nu extra deelnemer toe:
```
Welke atleet wil je toevoegen?
Yves
Welke tijd heeft hij/zij gelopen?
11,8

```
Bereken opnieuw de winnaar:
```
en het goud gaat naar ...: Yves
```





````python
class LoopWedstrijd:

    def __init__(self, naam, deelnemers = [], looptijden = []) -> None:
        self.naam = naam
        self.deelnemers = deelnemers
        self.looptijden = looptijden

    def geef_winnaar(self):
        kortste_tijd = min(self.looptijden)
        winnaar = self.deelnemers[self.looptijden.index(kortste_tijd)]
        return winnaar
    
    def voeg_prestatie_toe(self, naam, tijd):
        if naam not in self.deelnemers:
            self.deelnemers.append(naam)
            self.looptijden.append(float(tijd))
        else:
            self.looptijden[self.deelnemers.index(naam)] = tijd

    def __str__(self):
        deelnemers = '; '.join(self.deelnemers)
        tijden = '; '.join(map(str, self.looptijden))
        return f"deelnemers: {deelnemers[:-2]} \ntijden: {tijden}"
    

wedstrijd = LoopWedstrijd("OdiseeRun", ["Peter", "Kristien", "Thomas", "Evert-Jan"], [12.5, 14.8, 13.2, 14.5])
print(wedstrijd)
print(wedstrijd.geef_winnaar())

naam = input("Welke atleet wil je toevoegen?")
tijd = input("Welke tijd heeft hij/zij gelopen?")
wedstrijd.voeg_prestatie_toe(naam, tijd)


print(wedstrijd)
print(wedstrijd.geef_winnaar())

````
